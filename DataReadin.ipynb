```python
weball_header = list(pd.read_excel(r"C:\Users\krred\Documents\Berkeley\w209\Final_Project\weball_header.xlsx").columns)
```

```python
donations = pd.DataFrame()

for yr in ["08","10","12","14","16","18"]:
    temp = pd.read_csv(r"C:\Users\krred\Documents\Berkeley\w209\Final_Project\weball{}.txt".format(yr), sep = "|" , names = weball_header)
    temp['year'] = "20" + yr
    temp['year'] = pd.to_numeric(temp['year'])
    temp['lastname'] = temp['CAND_NAME'].str.split(",", expand = True)[0].str.lower().str.strip()
    temp['state_po'] = temp["CAND_ID"].str[2:4]
    temp['party_abb'] = temp['CAND_PTY_AFFILIATION'].str[0].str.lower()
    donations = donations.append(temp)
donations
```

```python
votes = pd.read_csv(r"C:\Users\krred\Documents\Berkeley\w209\Final_Project\1976-2018-house.csv", encoding = 'latin1')
votes['temp'] = np.where(votes['candidate'].isna(),"",votes['candidate'])
votes['lastname'] = votes['temp'].apply(lambda x: HumanName(x).last)
votes['lastname'] = votes['lastname'].str.lower()
votes['party_abb'] = votes['party'].str[0].str.lower()
votes
```

```python
house_votes_donations = pd.merge(votes,donations,on = ['lastname','party_abb','state_po','year'], how = 'inner')
house_votes_donations
```
